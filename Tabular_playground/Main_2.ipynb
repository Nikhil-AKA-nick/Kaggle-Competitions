```python
df = df.drop("id",axis =1)
```

```python
cols = ["product_code","attribute_0","attribute_1"]
from sklearn.preprocessing import LabelEncoder
le = LabelEncoder()
for col in cols:
    df[col] = le.fit_transform(df[col])
```

```python
df["measurement_3"].mean()
```

```python
df["measurement_3"].isnull()
```

```python
df["measurement_3"] = df["measurement_3"].fillna(df["measurement_3"].mean())
```

```python
cols_to_scale = ["loading","measurement_4","measurement_5","measurement_6","measurement_7","measurement_8","measurement_9","measurement_10","measurement_11","measurement_12","measurement_13","measurement_14","measurement_15","measurement_16","measurement_17"]
for col in cols_to_scale :
    mean = df[col].mean()
    df[col] = df[col].fillna(mean)
```

```python
df.info()
```

```python
X = df.drop("failure",axis = 1)
y = df["failure"]
```

```python
# from sklearn.model_selection import train_test_split
# X_train, X_test, y_train, y_test = train_test_split(X,y,test_size=0.2,random_state = 10)
```

```python
from sklearn.linear_model import LogisticRegression
lr_model = LogisticRegression()
```

```python
lr_model.fit(X,y)
```

```python
test_df = pd.read_csv("test.csv")
test_ids = test_df["id"]
```

```python
test_df.head()
```

```python
test_df = test_df.drop("id", axis = 1)
test_df.head()
```

```python
cols = ["product_code","attribute_0","attribute_1"]
from sklearn.preprocessing import LabelEncoder
le = LabelEncoder()
for col in cols:
    test_df[col] = le.fit_transform(test_df[col])
```

```python
cols_to_scale = ["loading","measurement_3","measurement_4","measurement_4","measurement_5","measurement_6","measurement_7","measurement_8","measurement_9","measurement_10","measurement_11","measurement_12","measurement_13","measurement_14","measurement_15","measurement_16","measurement_17"]
for col in cols_to_scale :
    mean = test_df[col].mean()
    test_df[col] = test_df[col].fillna(mean)
```

```python
test_df.info()
```

```python
Submission_prediction = lr_model.predict(test_df)
```

```python
df = pd.DataFrame({"id" : test_ids.values,
                  "failure": Submission_prediction,})
```

```python
df.to_csv("submission.csv", index = False)
```
